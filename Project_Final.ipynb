# Data Description: Calculation of Stats of the DataFrame 

```python
import numpy as np
def get_stats(x):
    frame_dict = {'Column':[], 'Non-null counts':[], 'Null values':[], 'Number of unique':[], 'Unique values':[], 'Data type':[]}
    for col in x.columns:
        frame_dict['Column'].append(col)
        frame_dict['Non-null counts'].append(x[col].count())
        frame_dict['Null values'].append(x[col].isnull().sum())
        frame_dict['Number of unique'].append(x[col].nunique())
        
        un_col = x[col].unique()
        
        if(len(un_col) > 4):
            frame_dict['Unique values'].append(np.append(un_col[:4], '...'))
        else:
            frame_dict['Unique values'].append(un_col)
        
        frame_dict['Data type'].append(x[col].dtype)
        
    return pd.DataFrame(frame_dict)



get_stats(df)
```

# Visualization 1 : No. of Flights per Carrier

```python
#No. of Flights per Carrier
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 8))
bands_country = df['Airline_Carrier'].value_counts()
plt.title('No. of flights per carrier')
sns.barplot(x=bands_country[:12].keys(),
            y=bands_country[:12].values, palette="GnBu_d")
plt.savefig('ac.png')
```

# Visualization 1.1 : No. of Flights per Airport

```python
#No. of Flights per Airport
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 8))
bands_country = df['Airport_Departure_Code'].value_counts()
plt.title('No. of flights per Airport')
sns.barplot(x=bands_country[:18].keys(),
            y=bands_country[:18].values, palette="GnBu_d")
plt.savefig('ac1.png')
```

# Visualization 2 : Delays by Airport

```python
#Delays by Airport
airport_by_delay = df.groupby('Airport_Departure_Code').Delay_Time_Arrival.sum().sort_values(ascending=False)
plt.figure(figsize=(20, 6))
airport_by_delay[:15].plot.bar()
plt.title('Delays by Airport')
plt.xlabel('Airport')
plt.ylabel('Hours')
plt.savefig('dap.png')

plt.show()
```

# Visualization 2.1 : Delays by Flight_Number

```python
#Delays by Airport
airport_by_delay = df.groupby('Flight_Number').Delay_Time_Arrival.sum().sort_values(ascending=False)
plt.figure(figsize=(20, 6))
airport_by_delay[:15].plot.bar()
plt.title('Delays by Flight Number')
plt.xlabel('Airport')
plt.ylabel('Hours')
plt.savefig('dap1.png')

plt.show()
```

# Visualization 3 : Delayed flights per Carrier

```python
#Delayed flights per Carrier
df['Delayed'] = df.loc[:,'Delay_Time_Arrival'].values > 0
df = df.sort_values(['Delayed']).reset_index(drop=True)
figsize=plt.subplots(figsize=(10,12))
sns.countplot(x='Delayed',hue='Airline_Carrier',data=df)
plt.title("Delayed Status")
plt.savefig("delayed.png")
plt.show()
```
